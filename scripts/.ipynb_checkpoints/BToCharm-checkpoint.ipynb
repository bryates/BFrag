## Define $\ttbar$ region

```python
def is_ttbar(jets, leptons):
        jets_pt = ak.fill_none(ak.pad_none(jets.pt, 1), 0)
        leps_pt  = ak.fill_none(ak.pad_none(leptons.pt, 1), 0)
        
        jets_eta = ak.fill_none(ak.pad_none(jets.eta, 1), 0)
        leps_eta  = ak.fill_none(ak.pad_none(leptons.eta, 1), 0)
        
        nJets = ak.num(events.Jet)
        print(nJets)
        nLep = ak.num(leptons)
        jpt30 = jets_pt[:,0] > 30
        lpt25 = (leps_pt[:,0] > 25)
        jeta24 = abs(jets_eta) < 2.4
        leta24 = abs(leps_eta) < 2.4
        #is_ttbar = (ak.num(jets)>4)&(ak.num(bjets_tight)>1)&((ak.num(ele)>1)|(ak.num(mu)>1))# & (pad_jets[:,0].pt>30)# & ((pad_ele[:,0].pt>25) | (pad_mu[:,0].pt>25))#&(ht>180)
        return (nJets >= 1) & (nLep >= 1) & jpt30 & lpt25
```

## Define some fit functions for fitting the $\Dz$ mass peak (Gaussian + exponential background)

```python
def d0_mass_fit(mass, mean, sigma):
    return(1/np.sqrt(2 * np.pi * np.square(sigma)) * np.exp(-(np.square(mass - mean) / np.square(sigma))))
```

```python
def expo_bkg(mass, l):
    return(np.exp(-l * mass))
```

## Fit the mass peak

```python
chi2_mask = (charm_cand.chi2 < 5) & (charm_cand.svprob>0.02)
b_mask = jets[charm_cand.jetIdx].btagDeepFlavB > tight
d0 = charm_cand[chi2_mask & b_mask & events.is_ttbar]
#plt.hist(ak.fill_none(ak.firsts(d0.fit_mass)[events.is_ttbar], -1), np.linspace(1.7, 2.0, 31))
from scipy.optimize import curve_fit
#h,_ = np.histogram(ak.fill_none(ak.firsts(d0.fit_mass)[events.is_ttbar], -1), mass_bins)
h,_ = np.histogram(ak.fill_none(ak.firsts(d0.fit_mass), -1), mass_bins)
plt.hist(ak.fill_none(ak.firsts(d0.fit_mass), -1), mass_bins, label='Data')
#m = [d0_mass_fit(mass=x, mean=1.87, sigma=.01, nsig=650, l=-2.27, nbkg=2e5, mean_kk=1.77, sigma_kk=.02, nkk=0, mean_pp=1.94, sigma_pp=.02, npp=0) for x in mass_bins]
m = [d0_mass_fit(mass=x, mean=1.87, sigma=.01, nsig=65, l=-2.27, nbkg=35481) for x in mass_bins]
plt.step(mass_bins, m, label='Initial guess')
#popt, pcov = curve_fit(d0_mass_fit, mass_bins[:-1], h, p0=[1.87,.01, 650, -2.27, 2e5, 1.77, .02, 1, 1.94, .02, 1])
d0_mean0, d0_sigma0, nd0, l0, ne0 = 1.87,.01, 650, -2.27, 2e5
popt, pcov = curve_fit(d0_mass_fit, mass_bins[:-1], h, p0=[d0_mean0, d0_sigma0, nd0, l0, ne0])
plt.plot(mass_bins, d0_mass_fit(mass_bins, *popt), label='Fit')
plt.legend()
print({name: round(val,2) for name,val in zip(('D0 mean', 'D0 width', 'N D0', 'lambda', 'N bkg'), popt)})
print(f'N D0 {round(popt[2])} +/- {round(np.sqrt(pcov[2][2]))}, N bkg {round(popt[4])} +/- {round(np.sqrt(pcov[4][4]))}')
#print(pcov)
```

```python
binned_masses = [d0.fit_mass[(xb > low) & (xb < high)] for low,high in zip(xb_bins[:-1], xb_bins[1:])]
d0_mean0, d0_sigma0, nd0, l0, ne0 = 1.87,.01, 1, -2.27, 30
d0mb, d0sb, ndb, lb, neb = (1.7, 2.0), (0, 1), (0, 1000), (-10, 10), (0, 100000)
for i,m in enumerate(binned_masses):
    if ak.num(m, axis=0) < 2:
        continue
    h,_ = np.histogram(ak.flatten(m), mass_bins)
    plt.hist(ak.flatten(m), mass_bins, label=f'D0 {i}')
    g = [d0_mass_fit(x, d0_mean0, d0_sigma0, nd0, l0, ne0) for x in mass_bins]
    #plt.step(mass_bins, g, label=f'Initial guess {i}')
    popt, pcov = curve_fit(d0_mass_fit, mass_bins[:-1], h, p0=[d0_mean0, d0_sigma0, nd0, l0, ne0], bounds=([1.7, 0, 0, -10, 0], [2.0, 1, 1000, 10, 10000]))
    plt.plot(mass_bins, d0_mass_fit(mass_bins, *popt), label=f'Fit {i}')
    #print(*popt)
    print({name: round(val,2) for name,val in zip(('D0 mean', 'D0 width', 'N D0', 'lambda', 'N bkg'), popt)})
    print(f'N D0 {round(popt[2])} +/- {round(np.sqrt(pcov[2][2]))}, N bkg {round(popt[4])} +/- {round(np.sqrt(pcov[4][4]))}')
plt.legend()
```
